```python
# import cell
import numpy as np
import pandas as pd
from os import walk
import os

from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
# load the 
df = pd.read_csv('Data/tfidf_usa.csv').set_index(['Year', 'ISO-alpha3 Code'])
```

```python
# get the words from the dataframe
words = np.array(df.columns)

# iterate over the indices
for index in df.index:
    
    # get the word counts of one document
    counts = np.array(df.loc[index])
    
    # create a dict with the 'word: count' structure 
    frequencies = dict(zip(words, counts))
    
    # create a wordcloud from the frequencies dict
    wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").fit_words(frequencies)
    
    # create a figure and save it using the year and ISO-alpha3 code from the dataframe index
    fig = plt.figure()    
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    filename = index[1] + str(index[0]) + '.png'
    fig.savefig('Wordclouds/' + filename)
    
```
